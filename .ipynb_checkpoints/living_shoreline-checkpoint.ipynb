```python
## Importing data from American Community Survey for Per Capita Income(future work, left on 07/10/23)
## https://www.nyc.gov/site/planning/data-maps/open-data/districts-download-metadata.page
```

```python
## Importing poverty data from American Community Survey(Census data)
file_path = "/Users/kenneth/Documents/Living_Shorelines/data/census_tracts_2020/socioeconomic_data/nyc_poverty_data_ACS_2020.csv"
nyc_poverty = pd.read_csv(file_path, low_memory=False)
```

```python
## cleaning the data so that we only have GEO_ID and poverty data columns
## Dropping first row of data as it is metadata/extra titles 

nyc_poverty = nyc_poverty[['GEO_ID','S1701_C01_001E']]
nyc_poverty = nyc_poverty.iloc[1: , :]
nyc_poverty['S1701_C01_001E'] = nyc_poverty['S1701_C01_001E'].astype(int)
nyc_poverty.info()
```

```python
## looking a New York City population data at the census tract resolution 
## Using American Community Survey Table: S0101
## Importing poverty data from American Community Survey(Census data)

file_path = "/Users/kenneth/Documents/Living_Shorelines/data/census_tracts_2020/socioeconomic_data/nyc_population_data_2020.csv"
nyc_pop = pd.read_csv(file_path, low_memory=False)
```

```python
## cleaning the data so that we only have GEO_ID and population data columns
## Dropping first row of data as it is metadata/extra titles 

nyc_pop = nyc_pop[['GEO_ID', 'S0101_C01_001E']] 
nyc_pop = nyc_pop.iloc[1: , :]
nyc_pop['S0101_C01_001E'] = nyc_pop['S0101_C01_001E'].astype(int)
nyc_pop.info()
                  
```

```python
nyc_pop.max()
```

```python
## LETS MERGE THESE DATASETS!!
nyc_poverty_ratio_test =  pd.merge(
    left = nyc_pop,
    right = nyc_poverty,
    how="inner",
    on='GEO_ID')
```

```python
nyc_poverty_ratio_test.plot()
```
